#Codigo para la información de los cuadros



```python
import pandas as pd

# Cargar el archivo de Excel
xls = pd.ExcelFile(r"Tabla MPI.xlsx")

# Leer las hojas del archivo de Excel en dataframes
df1 = pd.read_excel(xls, 'National MPI Results')
df2 = pd.read_excel(xls, 'Censored Headcounts')

#Unir las hojas en una sola base
df_unido = pd.merge(df1, df2, on=['ISO\ncountry code', 'Country', 'World region', 'Multidimensional Poverty Index\n(MPI = H*A) (Range 0 to 1)', 'Total number of indicators included \n(out of ten)','Indicator (s) missing'])

#renombrar las columnas 
df_unido = df_unido.rename(columns={
    "ISO\ncountry code": "Código del país",
    "Country": "País",
    "World region": "Región",
    "Multidimensional Poverty Index\n(MPI = H*A) (Range 0 to 1)": "IPM (0 a 1)",
    "(H)  Headcount ratio: Population in multidimensional poverty\n(% Population)": "(H) Incidencia de la pobreza",
    "(A) Intensity of deprivation among the poor\n (Average % of weighted deprivations)": "(A) Intensidad de privación",
    "Population Year of the survey (Thousands)": "Población Año de la encuesta (Miles)",
    "Population 2019 (Thousands)": "Población 2019 (Miles)",
    "Population 2020 (Thousands)": "Población 2020 (Miles)",
    "MPI poor People Population Year of the survey (Thousands)": "Población pobre Año de la encuesta (Miles)",
    "MPI poor People Population 2019 (Thousands)": "Población pobre 2019 (Miles)",
    "MPI poor People Population 2020 (Thousands)": "Población pobre 2020 (Miles)",
    "Total number of indicators included (out of ten)": "Indicadores incluidos (de diez)",
    "Indicator (s) missing": "Indicador(es) faltante(s)",
    " Poor and deprived in Nutrition (% Population)": "Nutrición (%)",
    "Poor and deprived in Child mortality (% Population)": "Mortalidad infantil (%)",
    "Poor and deprived in Years of schooling (% Population)": "Años de escolaridad (%)",
    "Poor and deprived in School attendance (% Population)": "Asistencia escolar (%)",
    "Poor and deprived in Cooking \nfuel (% Population)": "Combustible para cocinar (%)",
    "Poor and deprived in Sanitation (% Population)": "Salubridad (%)",
    "Poor and deprived in Drinking water (% Population)": "Agua potable (%)",
    "Poor and deprived in Electricity (% Population)": "Electricidad (%)",
    "Poor and deprived in Housing (% Population)": "Vivienda (%)",
    "Poor and deprived in Assets (% Population)": "Activos (%)"
})

df_unido.head()
```

```python
##Cuadro 1
import pandas as pd
# Se calculan el total de la población por región y año
total_poblacion_2019 = df_unido.groupby('Región')['Población 2019 (Miles)'].sum()/1000
total_poblacion_2020 = df_unido.groupby('Región')['Población 2020 (Miles)'].sum()/1000


# Se calcula el total de la población pobre por región y año
total_pobre_2019 = df_unido.groupby('Región')['Población pobre 2019 (Miles)'].sum()/1000
total_pobre_2020 = df_unido.groupby('Región')['Población pobre 2020 (Miles)'].sum()/1000


#Se calcula el porcentaje de población pobre por región y año
porcentaje_pobre_2019 = (total_pobre_2019 / total_poblacion_2019) * 100
porcentaje_pobre_2020 = (total_pobre_2020 / total_poblacion_2020) * 100


# Se crea un nuevo DataFrame con los resultados
resumen = pd.DataFrame({
    'Poblacion 2019 (Millones)': total_poblacion_2019,
    'Poblacion 2020 (Millones)': total_poblacion_2020,

    'Población pobre segun el IPM 2019(Millones)': total_pobre_2019,
    'Población pobre segun el IPM 2020(Millones)': total_pobre_2020,


    'Porcentaje de población pobre según IPM 2019(Millones)': porcentaje_pobre_2019,
    'Porcentaje de población pobre según IPMPobre 2020(Millones)': porcentaje_pobre_2020,

})
#Se convierte la tabla en un excel
resumen.to_excel('Cuadro1.xlsx')
print(resumen)
```

```python
import pandas as pd


columnas = ['Nutrición (%)',
       'Mortalidad infantil (%)',
       'Años de escolaridad (%)',
       'Asistencia escolar (%)',
       'Combustible para cocinar (%)',
       'Salubridad (%)',
       'Agua potable (%)',
       'Electricidad (%)',
       'Vivienda (%)',
       'Activos (%)'] 

# Se seleccionan las columnas deseadas
df_unido2 = df_unido[columnas]

#Se calculan las estadísticas descriptivas
estadisticas = df_unido2.describe()

estadisticas_transpuestas = estadisticas.T
#Se convierte la tabla a excel
estadisticas_transpuestas.to_excel('Cuadro2.xlsx')
```

```python

# Se agrupa por 'Región' y calculamos el promedio y la desviación estándar
result = df_unido.groupby('Región').agg({'IPM (0 a 1)': ['mean', 'std'], '(H) Incidencia de la pobreza': ['mean', 'std'], '(A) Intensidad de privación': ['mean', 'std']})

print(result)
#Se convierte la tabla a excel
result.to_excel('Cuadro3.xlsx')
```
